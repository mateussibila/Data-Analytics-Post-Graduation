```python
# Salve todos os IDs dos pedidos com pelo menos um valor ausente em 'add_to_cart_order'

# Criando um novo dataframe onde foram filtrados as linhas com valores ausentes
# em 'add_to_cart_order' e extraindo apenas a coluna 'order_id'
order_products_missing_values = order_products[order_products['add_to_cart_order'].isna()]['order_id']

print(order_products_missing_values)
```

```python
# Todos os pedidos com valores ausentes contêm mais de 64 produtos?
# Agrupe os pedidos com dados ausentes por ID de pedido
# Conte o número de 'product_id' em cada pedido e verifique o valor mínimo da contagem

# Filtrando o dataframe original apenas com 'order_id' que tenham
# ao menos um valor ausente na coluna 'product_id'
order_products_filtered = order_products[order_products['order_id'].isin(order_products_missing_values)]
print(order_products_filtered)
print()

# Agrupando o dataframe filtrado 'order_products_filtered' por 'order_id'
order_products_grouped = order_products_filtered.groupby('order_id')['product_id'].count()

# Ordenando os valores em ordem crescente e imprimindo
print(order_products_grouped.sort_values())
```

- Fiz o grupamento desse novo dataframe filtrado, agrupando por 'order_id', e contando os valores da coluna 'product_id' para saber o tamanho desses pedidos (quantidade de itens). O tamanho do pedido pode ter ocasionado um erro de alguma forma, pois todos os 70 pedidos que tem valores ausentes na coluna 'add_to_cart_order' são pedidos "grandes", com mais de 64 itens.

### [C2] Quais são os 20 principais itens incluídos com mais frequência em pedidos repetidos? Exiba os IDs e nomes.

```python
# Criando um novo dataframe, agrupando por produto e somando quantas vezes se repetem
order_products_rep_sum = order_products.groupby('product_id').agg({'reordered':'sum'})

# Renomeando a coluna 'reordered' para refletir a nova informação (soma)
order_products_rep_sum.rename(columns={'reordered': 'reordered_sum'}, inplace=True)

# Aplicando ordem decrescente
order_products_rep_sum.sort_values(by='reordered_sum', ascending=False, inplace=True)
```

```python
# Criando um novo dataframe com os top 20
order_products_rep_sum_top_20 = order_products_rep_sum.head(20)

# Realizando o merge com o dataframe products
order_products_rep_sum_top_20 = order_products_rep_sum_top_20.merge(products, on='product_id', how='left')

# Removendo as colunas desnecessárias
order_products_rep_sum_top_20.drop(['aisle_id', 'department_id'], axis='columns', inplace=True)

# Imprimindo
print(order_products_rep_sum_top_20)
```

- Para conhecer os itens incluídos com mais frequência em pedidos repetidos precisamos contar quantas vezes cada pedido aparece com o valor 1 na coluna 'reordered', o que pode ser feito aplicando um método soma, já que essa coluna sempre terá o valor 1 ou 0.
    
- Após esta operação, reordenamos e extraímos os top 20 produtos mais 'reordenados'. Realizamos um merge com outro dataframe para adicionar a informção 'nome do produto'.


### [C3] Para cada produto, qual parcela de todos os pedidos dele são repetidos?

```python
# Criando um novo dataframe para calcular a taxa de repetição
# Taxa de repetição = (Total de vezes que o produto foi pedido) / (Vezes em que era uma repetição, ou seja, reordered = 1)
order_products_rep_rate = order_products_count.merge(order_products_rep_sum, on='product_id', how='outer')

# Removendo colunas desnecessárias
order_products_rep_rate.drop(['add_to_cart_order', 'reordered'], axis='columns', inplace=True)

# Criando uma nova coluna que represente a proporção entre as outras duas
order_products_rep_rate['repetiton_rate'] = order_products_rep_rate['reordered_sum'] / order_products_rep_rate['order_count']

# Realizando um merge para adicionar o nome do produto
order_products_rep_rate = order_products_rep_rate.merge(products, on='product_id', how='left')

# Removendo as colunas desnecessárias
order_products_rep_rate.drop(['aisle_id', 'department_id'], axis='columns', inplace=True)

print(order_products_rep_rate)
```

Escreva suas conclusões aqui

<span style="color: blue;">
    
- Para descobrir com que frequência cada produto se repete, precisamos comparar quantas vezes um produto aparece no total de todos os pedidos versus quantas vezes aquela compra foi um caso de 'reordered' = 1, ou seja, de repetição.
    
- Já tínhamos dois dataframes com as informações necessárias e com esses dados pudemos calcular com uma divisão de um valor pelo outro, ou seja (repetições) dividido por (total de aparições).
    
- Ao final obtivemos o fator 'repetiton_rate', que varia de 0 a 1 e representa uma proporção de repetição para cada produto.
    


### [C4] Para cada cliente, qual proporção de todos os seus pedidos são repetidos?

- Na sequência, agrupamos dos dados por usuário, fazendo uma média da coluna 'reordered', o que resulta em uma proporção que representa de 0 a 1, quantos dos produtos comprados por aquele usuário foram comprados mais de uma vez, obtendo assim o resultado final.
